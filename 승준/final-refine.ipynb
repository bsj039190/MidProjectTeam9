```python
DATA_PATH = '/content/drive/MyDrive/멋사/mid project/'
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(DATA_PATH + 'first_refine.csv')
df.head()
```

```python
org_df = df.copy()
```

- object 타입인 공연날짜를 datetime으로 변경

```python
org_df['공연날짜'] = pd.to_datetime(org_df['공연날짜'])
org_df.head()
```

- 공연시간을 포매팅해서 공연시간_formatted에 삽입

```python
org_df['공연시간_str'] = org_df['공연시간'].astype(str).str.zfill(6)  # 6자리로 맞추기
org_df['공연시간_formatted'] = org_df['공연시간_str'].str[:2] + ':' + org_df['공연시간_str'].str[2:4] + ':' + org_df['공연시간_str'].str[4:]
org_df.head()
```

```python
org_df['공연날짜'] = pd.to_datetime(org_df['공연날짜'].dt.date.astype(str) + ' ' + org_df['공연시간_formatted'])
org_df.head()
```

```python
org_df = org_df.drop(columns=['공연시간_str', '공연시간_formatted', '공연시간'])
```

```python
org_df.head()
```

# 구매날짜, 구매시간도 똑같이 진행

```python
org_df.info()
```

```python
org_df['구매날짜'] = pd.to_datetime(org_df['구매날짜'], format='%Y%m%d', errors='coerce')
org_df.head()
```

```python
org_df['구매시간_str'] = org_df['구매시간'].astype(str).str.zfill(6)  # 6자리로 맞추기
org_df['구매시간_formatted'] = org_df['구매시간_str'].str[:2] + ':' + org_df['구매시간_str'].str[2:4] + ':' + org_df['구매시간_str'].str[4:]
org_df.head()
```

```python
org_df['구매날짜'] = pd.to_datetime(org_df['구매날짜'].dt.date.astype(str) + ' ' + org_df['구매시간_formatted'])
org_df.head()
```

```python
org_df = org_df.drop(columns=['구매시간_str', '구매시간_formatted', '구매시간'])
org_df.head()
```

```python
org_df['총판매석'] = org_df['총판매석'].fillna(0).astype(int)
org_df['일반판매석'] = org_df['일반판매석'].fillna(0).astype(int)
org_df['장애인석'] = org_df['장애인석'].fillna(0).astype(int)
org_df['OP석'] = org_df['OP석'].fillna(0).astype(int)
org_df['판매유보석'] = org_df['판매유보석'].fillna(0).astype(int)
org_df.info()
```

```python
org_df.head()
```

```python
org_df = org_df.drop(columns=['OP석'])
```

```python
org_df.head()
```

```python
file_save = DATA_PATH + 'refine_result.csv'
org_df.to_csv(file_save, index=False)
```
