```python
df.columns = ['공연명','공연장소명','공연일자','공연시간','장르명','가격내용','홈페이지URL','결제일자','결제시간','할인내용','결제금액','회원등급명','나이값','성별값','거주지역명','환불여부']
```

```python
df.drop(['가격내용','홈페이지URL'], axis=1, inplace=True)
```

```python
import numpy as np
```

```python
df["성별값_F"] = df["성별값"] == "F"
df["환불여부_Y"] = df["환불여부"] == "Y"
```

```python
df['성별값_F'] = df['성별값_F'].astype(int)
df['환불여부_Y'] = df['환불여부_Y'].astype(int)
```

```python
df['나이값'] = df['나이값'].str.extract('(\d+)')
```

```python
df_filtered = df[df['환불여부_Y'] != 1]

print(df_filtered)
```

```python
booking_count_by_time = df['공연시간'].value_counts()

total_bookings = len(df['공연시간'])

booking_ratio_by_time = (booking_count_by_time / total_bookings) * 100

print(booking_ratio_by_time)
```
